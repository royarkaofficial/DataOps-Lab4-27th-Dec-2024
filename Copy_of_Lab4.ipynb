```python
fig , ax = plt.subplots(2,2,figsize=(10,10))
# 10 inches by 10 inches
#ax[2,2]
#ax[0,0] - Top-left
#ax[0,1] - Top-right
#ax[1,0] - Bottom-left
#ax[1,1] - Bottom-right
ax[0,0].plot(["Bhandaru","Nagamani","Bharadwaj"],[23,25,23])
ax[0,0].set_title("name_age_graph")
ax[0,0].set_xlabel("Name")
ax[0,0].set_ylabel("Age")

ax[0,1].plot(["Trainer","Architect","Manager"],[25,23,28])
ax[0,1].set_title("Designation_age_graph")

ax[1,0].plot(["35000","48000","50000"],[23,25,28])
ax[1,0].set_title("Salary_age_graph")

ax[1,1].plot(["Tirupati","Hindupur","Hyderabad"],[28,23,26])
ax[1,1].set_title("City_age_graph")

plt.tight_layout()
plt.show()
```

Basic Plot types in MatplotLib

```python
#Line plot
plt.plot([1,2,3,4],[2,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[4,15,16,22])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot example")
plt.show()
```

```python
#Scatter plot
plt.scatter(['Cat','Dog','Lion','Cheetah'],[1,15,6,20])
plt.xlabel("Animals")
plt.ylabel("Age")
plt.title("Scatter Plot example")
plt.show()
```

```python
#Bar plot
products = ["Machinary","Cosmetics","Electronics","Carpentary"]
price = [38500 , 50000 , 48521 , 35000]

plt.bar(products,price)
plt.xlabel("Products")
plt.ylabel("Price")
plt.title("Bar Plot example")
plt.show()
```

```python
#Histograms
scores = [88,92,45,57,66,98]
plt.hist(scores)
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram example")
plt.show()
```

```python
#Pie plot

comp = ["Google","Microsoft","CloudThat","Mphasis"]
share = [15,25,5,20]
plt.pie(share,labels=comp, autopct = "%1.1f%%") #25.0%
plt.title("Pie Plot example")
plt.show()
```

```python
#Heatmap
import numpy as np

data = np.random.rand(10,10)
plt.imshow(data, cmap="hot")
plt.colorbar()
plt.title("Heat map example")
plt.show()
```

```python
#Animations in MatplotLib
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
fig , ax = plt.subplots()
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

line , = ax.plot(x,y , color = "red")
```

```python
def animate(frame):
  line.set_ydata(np.sin(x+0.1*frame))
  return line ,
  ani = FuncAnimation(fig , animate , frames = 10 , interval = 200 , blit = True)
  plt.show()
  ani.save("animation.gif")
```

Create a Scatter plot animation

```python
fig , ax = plt.subplots()
x_data = np.random.rand(10)
y_data = np.random.rand(10)
scat = ax.scatter(x_data,y_data)

def update(frame) :
  new_x = np.random.rand(10)
  new_y = np.random.rand(10)
  scat.set_offsets(np.c_[new_x,new_y])
  return scat ,
#set_offsets : Updating the position of scatter plot points

ani = FuncAnimation(fig , update , frames = 10 , interval = 100 , blit = True)
plt.show()
ani.save("scatter_animation.gif")
```
